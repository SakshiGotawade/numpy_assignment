# Step 1 : IMPORTING PANDAS

```python
import pandas as pd
```

# STEP 2: CREATING PANDAS

```python
data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }
```

  # 2a. Print the first five rows of the Data Frame

```python
df = pd.DataFrame(data)
print (df.head())
```

# 2b. Get summary statistics for Age and Salary

```python
print(df[['Age', 'Salary']].describe())
```

# 2c: Average salary of HR department

```python
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print(hr_avg_salary)
```

# 2. Adding Bonus column (10% of Salary)

```python
df['Bonus'] = df['Salary'] * 0.10
print (df)
```

# 3. Filtering the DataFrame to show employees aged between 25 & 30

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print (filtered_df)
```

# 4. Average salary for each department


```python
avg_salary_by_dept = df.groupby('Department')['Salary'].mean()
print(avg_salary_by_dept)
```

# 5. Sorting the DataFrame by Salary in ascending order & # Save the sorted DataFrame to a CSV file

```python
sorted_df = df.sort_values(by='Salary', ascending=True)
sorted_df.to_csv('sorted_salary.csv', index=False)
print(sorted_df)
```
